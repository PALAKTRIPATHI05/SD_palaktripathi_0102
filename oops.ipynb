```python
class student:
  def __init__(self,id,name,marks):
    self.id=id
    self.name=name
    self.marks=marks
  def display(self):
    print('id:',self.id)
    print(self.name)
    print(self.marks)
s1= student(101,'palak',97)
s1.display()
```

inheritance in oops

```python
class parent:
  def test_parent(self):
    print('this is my parent class')
```

```python
class child(parent):
  pass
```

```python
child_obj=child()
```

```python
child_obj.test_parent()
```

```python
class class1:
  def test_class1(self):
    print('this is my class')
```

```python
class class2(class1):
  def test_class2(self):
    print('this is my class')
```

```python
class class3(class2):
  def test_class3(self):
    print('this is my class3')
```

```python
obj_class3= class3()
```

```python
obj_class3.test_class1()
```

```python
class class1:
  def test_class1(self):
    print('this is my class')
```

```python
class class2:
  def test_class2(self):
    print('this is my class')
```

```python
class class3(class1,class2):
  pass
```

```python
obj_class3=class3()
```

```python
obj_class3.test_class1()
```
