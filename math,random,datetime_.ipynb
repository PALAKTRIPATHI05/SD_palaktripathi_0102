```python
import time
from IPython.core.display import clear_output
while True:
  clear_output(wait=True)
  current=datetime.datetime.now().strftime('%H:%M:%S')
  print(current)
  time.sleep(1)
```

```python
import time
from zoneinfo import ZoneInfo
from IPython.core.display import clear_output
while True:
  clear_output(wait=True)
  current=datetime.datetime.now(ZoneInfo('Asia/Kolkata')).strftime('%H:%M:%S')
  print(current)


  time.sleep(1)
```
